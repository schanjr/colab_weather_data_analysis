```python
from IPython.core.display import HTML
HTML("<style>.container { width:100% !important; } </style>") #Makes the graphs fitting window size of the browser. 



def is_numeric(obj):
    attrs = ['__add__', '__sub__', '__mul__', '__div__', '__pow__']
    return all(hasattr(obj, attr) for attr in attrs)
 
#Example of MAP
def convert_dt(x):
    if is_numeric(x) and len(str(x))==10: #condition checks is the returned cell is numeric and has len of 10 so it guarantees we're converting date numbers only
        return pd.to_datetime(x, unit='s')
    elif is_numeric(x) and len(str(x))==9: #condition checks is the returned cell is numeric and has len of 10 so it guarantees we're converting date numbers only
        return pd.to_datetime(x, unit='s')
    else:
        return x
     

```

```python
new_df=daily_df_2000_2016.applymap(convert_dt)
new_df['averageTemperature']=(new_df['apparentTemperatureMax']+new_df['apparentTemperatureMin'])/2
```

```python
Year_2000=new_df['2000'].copy()
Year_2000.drop(Year_2000.index[0],inplace=True)
Year_2001=new_df['2001'].copy().drop_duplicates()
Year_2002=new_df['2002'].copy().drop_duplicates()
Year_2003=new_df['2003'].copy().drop_duplicates()
Year_2004=new_df['2004'].copy()
Year_2004.drop(Year_2004.index[0],inplace=True)
Year_2005=new_df['2005'].copy().drop_duplicates()
Year_2006=new_df['2006'].copy().drop_duplicates()
Year_2007=new_df['2007'].copy().drop_duplicates()
Year_2008=new_df['2008'].copy()
Year_2008.drop(Year_2008.index[0],inplace=True)
Year_2009=new_df['2009'].copy().drop_duplicates()
Year_2010=new_df['2010'].copy().drop_duplicates()
Year_2011=new_df['2011'].copy().drop_duplicates()
Year_2012=new_df['2012'].copy()
Year_2012.drop(Year_2012.index[0],inplace=True)
Year_2013=new_df['2013'].copy()
Year_2013.drop(Year_2013.index[0],inplace=True)
Year_2013.interpolate(inplace=True)
Year_2014=new_df['2014'].copy()
Year_2014.drop_duplicates()
Year_2015=new_df['2015'].copy().drop_duplicates().interpolate()
Year_2015=Year_2015.append(Year_2015.iloc[-1])



```

```python
def month_cat(num):
    num=int(num)
    if num==1: return 'Jan'
    if num==2: return 'Feb'
    if num==3: return 'Mar'
    if num==4: return 'Apr'
    if num==5: return 'May'
    if num==6: return 'Jun'
    if num==7: return 'Jul'
    if num==8: return 'Aug'
    if num==9: return 'Sep'
    if num==10: return 'Oct'
    if num==11: return 'Nov'
    if num==12: return 'Dec'

#new_df.drop('month_category', axis=1, inplace=True)
cat_list=list()
for i in agg_df.index.month:
    cat_list.append(month_cat(i))
agg_df['month_category']=cat_list
```

```python
from scipy import stats

total_p=0

base_year=2014
for num in xrange(2000,2016):
    t_stat, two_tail_p_value=stats.ttest_ind(agg_df["Year_{0}".format(base_year)],agg_df["Year_{0}".format(num)],equal_var=False)  #Conducts Welch's Test. Note that Pvalue response from stats.ttest_ind default is two tail. 
    print "Year {b} and Year {0}: T_Statistics:{T}    P-Value:{P}".format(num, b=base_year, T=t_stat, P=two_tail_p_value)
    total_p+=two_tail_p_value
    
print total_p
```
